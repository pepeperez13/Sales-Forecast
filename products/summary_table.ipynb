```python
import pandas as pd

# Load the CSV data
data = pd.read_csv(
    "/Users/pascalvenema/Documents/GitHub/AFCS/assignments/final-project/products/metrics.csv"
)

# Generate summary statistics
summary_table = {
    "Metric": ["MAE", "RMSE", "Test_Weekly_Average_Sales", "Test_Total_Sales"],
    "Mean": [
        data["MAE"].mean(),
        data["RMSE"].mean(),
        data["Test_Weekly_Average_Sales"].mean(),
        data["Test_Total_Sales"].mean(),
    ],
    "Median": [
        data["MAE"].median(),
        data["RMSE"].median(),
        data["Test_Weekly_Average_Sales"].median(),
        data["Test_Total_Sales"].median(),
    ],
    "Standard Deviation": [
        data["MAE"].std(),
        data["RMSE"].std(),
        data["Test_Weekly_Average_Sales"].std(),
        data["Test_Total_Sales"].std(),
    ],
    "Minimum": [
        data["MAE"].min(),
        data["RMSE"].min(),
        data["Test_Weekly_Average_Sales"].min(),
        data["Test_Total_Sales"].min(),
    ],
    "Maximum": [
        data["MAE"].max(),
        data["RMSE"].max(),
        data["Test_Weekly_Average_Sales"].max(),
        data["Test_Total_Sales"].max(),
    ],
}

summary_df = pd.DataFrame(summary_table)


summary_df
```
